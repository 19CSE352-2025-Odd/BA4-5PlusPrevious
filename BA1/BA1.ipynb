```python
import urllib.request
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
book_url = 'http://www.gutenberg.org/ebooks/42671.txt.utf-8'
```

```python
with urllib.request.urlopen(book_url) as f:
    Pride_and_Prejudice_book_text = f.read().decode('utf-8')
print(Pride_and_Prejudice_book_text)
```

```python
# Break the text into Chapters
Pride_and_Prejudice_book_text_book_chapters = Pride_and_Prejudice_book_text.split('CHAPTER ')
# Drop the first "Chapter" - it's the Project Gutenberg header
Pride_and_Prejudice_book_text_book_chapters = Pride_and_Prejudice_book_text_book_chapters[1:]
```

```python
Pride_and_Prejudice_book_text_df =pd.DataFrame(Pride_and_Prejudice_book_text_book_chapters, columns=['Chapters'])
```

```python
book_url = 'https://www.inferentialthinking.com/data/huck_finn.txt'
```

```python
with urllib.request.urlopen(book_url) as f:
    huck_finn_chapters_book = f.read().decode('utf-8')
```

```python
#Split by Chapter
huck_finn_chapters = huck_finn_chapters_book.split('CHAPTER ')
```

```python
#loading word count into array and into dataframe
Jim = 'Jim', np.char.count(huck_finn_chapters, 'Jim'),
Tom = 'Tom', np.char.count(huck_finn_chapters, 'Tom'),
Huck = 'Huck', np.char.count(huck_finn_chapters, 'Huck')
for i in Jim:
    #Jim_list = {'Jim':i}
    Jim_list = i
for i in Tom:
    Tom_list = i
for i in Huck:
    Huck_list = i
huck_finn_chapters_counts = pd.DataFrame({'Jim' : Jim_list,'Tom' : Tom_list, 'Huck' :Huck_list})
```

```python
huck_finn_chapters_counts
```

```python
#cumsum_df = counts['Jim'].cumsum(skipna = False)
huck_finn_chapters_cumsum_df = pd.DataFrame({'Jim_CumSum' : huck_finn_chapters_counts['Jim'].cumsum(skipna = False), \
                       'Tom_CumSum' : huck_finn_chapters_counts['Tom'].cumsum(skipna = False), \
                       'Huck_CumSum' :huck_finn_chapters_counts['Huck'].cumsum(skipna = False)})
```

```python
plt.plot( huck_finn_chapters_cumsum_df.index,huck_finn_chapters_cumsum_df['Jim_CumSum'],label=f'Jim', marker='',color='red')
plt.plot( huck_finn_chapters_cumsum_df.index,huck_finn_chapters_cumsum_df['Tom_CumSum'],label=f'Tom', marker='',color='blue')
plt.plot( huck_finn_chapters_cumsum_df.index,huck_finn_chapters_cumsum_df['Huck_CumSum'],label=f'Huck', marker='',color='Green')
plt.legend()
plt.title('Cumulative Number of Times Each Label Appears')
plt.xlabel('Chapters')
plt.ylabel('Word Count')
```

```python
# As we see see in the graph, Jim is a central character by the large number of times his name appears.
# Notice Tom is hardly mentioned for much of the book until he arrives and joins Huck and Jim
# Huck, his name hardly appears at all, because he is the narrator.
```

```python
little_women = 'http://www.gutenberg.org/cache/epub/514/pg514.txt'
```

```python
with urllib.request.urlopen(little_women) as f:
    little_women_book = f.read().decode('utf-8')
```

```python
#Split by Chapter
little_women_book_chapters = little_women_book.split('CHAPTER ')
```

```python
#loading word count into array and into dataframe
Amy = 'Amy', np.char.count(little_women_book_chapters, 'Amy'),
Beth = 'Beth', np.char.count(little_women_book_chapters, 'Beth'),
Jo = 'Jo', np.char.count(little_women_book_chapters, 'Jo')
Meg = 'Meg', np.char.count(little_women_book_chapters, 'Meg'),
Laurie = 'Laurie', np.char.count(little_women_book_chapters, 'Laurie')
for i in Amy:
    Amy_list = i
for i in Beth:
    Beth_list = i
for i in Jo:
    Jo_list = i
for i in Meg:
    Meg_list = i
for i in Laurie:
    Laurie_list = i
little_women_book_counts = pd.DataFrame({'Amy' : Amy_list,'Beth' : Beth_list, 'Jo' :Jo_list, 'Meg' : Meg_list, 'Laurie' : Laurie_list})
```

```python
little_women_book_counts
```

```python
#cumsum_df = counts['Jim'].cumsum(skipna = False)
little_women_book_cumsum_df = pd.DataFrame({'Amy_CumSum' : little_women_book_counts['Amy'].cumsum(skipna = False), \
                       'Beth_CumSum' : little_women_book_counts['Beth'].cumsum(skipna = False), \
                       'Jo_CumSum' :little_women_book_counts['Jo'].cumsum(skipna = False), \
                       'Meg_CumSum' :little_women_book_counts['Meg'].cumsum(skipna = False), \
                         'Laurie_CumSum' :little_women_book_counts['Laurie'].cumsum(skipna = False)})
```

```python
little_women_book_cumsum_df
```

```python
plt.plot( little_women_book_cumsum_df.index,little_women_book_cumsum_df['Amy_CumSum'],label=f'Amy', marker='',color='red')
plt.plot( little_women_book_cumsum_df.index,little_women_book_cumsum_df['Beth_CumSum'],label=f'Beth', marker='',color='blue')
plt.plot( little_women_book_cumsum_df.index,little_women_book_cumsum_df['Jo_CumSum'],label=f'Jo', marker='',color='green')
plt.plot( little_women_book_cumsum_df.index,little_women_book_cumsum_df['Meg_CumSum'],label=f'Meg', marker='',color='black')
plt.plot( little_women_book_cumsum_df.index,little_women_book_cumsum_df['Laurie_CumSum'],label=f'Laurie', marker='',color='orange')
plt.legend()
plt.title('Cumulative Number of Times Each Label Appears')
plt.xlabel('Chapters')
plt.ylabel('Word Count')
```
